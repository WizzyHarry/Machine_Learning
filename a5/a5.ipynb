## Regression Analysis

**Goal:** Find the best fitting line which can accurately predict the output.

**Fundamentals:**

*Dependent variables* (**output on y-axis**): always variables being studied--that is, whose variation(s) are being modified somehow!

*Independent variables* (**input on x-axis**): always variables being manipulated, to study and compare effects on dependent variables.

**Note:** The designations independent and dependent variables are used not to imply "cause and effect."

  1. Simplie linear regression (1 DV/1 IV)
  2. Multiple linear regression (1 DV/2 or more IVs)
  3. Multivariate linear regression (2 or more DVs/2 or more IVs)

**Simple linear regression:** Only **one** independent variable affecting **one** dependent variable.

**Multiple linear regression: Two or more** independent variables affecting **one** dependent variable.

**Multivariate linear regression: Two or more** independent variables affecting **two or more** dependent variable.

## Supervised vs. Unsupervised Learning

  - Supervised learning: Prior knowledge of what output values for samples should be.
  - Unsupervised learning: Does not have labeled outputs. Goal is to infer natural structure present within a set of data points.

## Create multiple regression model

### Multiple Regression Steps

  1. Identify x (IVs) and y (DV): x=Height,Width,VerticalLength, y=Weight
  2. Create Train and Test datasets
  3. Train model
  4. Evaluate model

```python
# focus on one species: Bream
bream = fish.query('Species == "Bream"')
```

```python
# identify IVs and DV
x = bream[['Height','Width','VerticalLength']]
y = bream[['Weight']]

# Note: Combination of 'height', 'width', 'verticallength', returns a higher correlation score than individual attributes
# Compare multiple regression model

# x = bream[['VerticalLength']]
# x = bream[['Width']]
# x = bream[['Height']]
```

```python
# Create train and test datasets
# a. split variables into training and testing sets
# b. build model using training, then run on testing

# Training data is the biggest subset of original dataset, used to train or fit the model
# Test dataset is another (smaller) subset or original data, independent from training set
# 75% training, 25% testing

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.75, test_size = 0.25, random_state = 100)
```

```python
# display x training dataset (IV)
x_train # in cm
```

```python
# display y training dataset (DV)
y_train/454 # convert weight in grams to LBs
```

## Fitting model: Find best fitting line -- to accurately predict output

```python
# fit regression line to plot
model = LinearRegression().fit(x_train, y_train)
```

```python
# display R2 correlation values--validates model through correlation score
# score() function: accepts test dataset and returns R2 correlation value
# higher score indicates better fit
model.score(x_test, y_test)
```

```python
# predict weight and display predicted values based upon IVs
y_predicted = model.predict(x_test)

# create predicted dataframe with "PredictedWeight" col using y_predicted values
predicted = pd.DataFrame(y_predicted, columns = ['PredictedWeight'])

# display data structure type and "predicted"
predicted
```

```python
# display data structure type and values
x_test # IVs
```

```python
# display data structure type and values
y_test # DV
```

```python
# join and display all cols, including PredictedWeight col
# pandas.DF.join(): join multiple dataframe objects by index
# Note: drop parameter removes old index vals being added as col
final = predicted.join([x_test.reset_index(drop=True),
                        y_test.reset_index(drop=True)])


# display predictedweight and actual weight (DVs) and IVs
final
```

## Residuals

  - R2 value returned by score() method for test dataset provides good indication of regression model validity.
  - Also, plotting residuals helps to evaluate models.

```python
# calculate and display residual values
# Note: "Residuals" are simply differences between DV test values and DV predicted values
# Residual = Actual - Predicted
final['Residual'] = final.Weight - final.PredictedWeight

final
```

```python
# plot residuals
# KDE (Kernel Density Estimate) plot: Estimates data distribution
# Visual representation of underlying distribution of data--helps to understand shape and spread of data
# Produces small continous curve (also called kernel) for every individual data point along an axis
# Scale of "density" axis depends on data values.

# Note: Here, most residuals w/in +100 and -100 of zero, and closely centered on zero (0)--indicating a decent model!
sns.kdeplot(data=final, x='Residual')
```
